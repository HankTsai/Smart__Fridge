```python
# 第二部分，清除食材名稱重複的部分
with open('./foodBank/recipe0928_V3.json', 'r', encoding='utf-8') as f:
    content = json.load(f)
```

```python
# 清除重複食材名稱
count = 0
for k in content:
    ck_ing = len(k['ingredient'])
    ck_sea = len(k['seasoning'])
    ck_set = set()
    index = []
    for i,j in enumerate(k['ingredient']):
        if j[0] not in ck_set:
            ck_set.add(j[0])
        else:    
            index.append(i)
    for n in range(1,len(index)+1):
        k['ingredient'].pop(index[-n])
    
    index = []
    for i,j in enumerate(k['seasoning']):
        if j[0] not in ck_set:
            ck_set.add(j[0])
        else:    
            index.append(i)
    for n in range(1,len(index)+1):
        k['seasoning'].pop(index[-n])
    
    if (len(k['ingredient']) != ck_ing) or (len(k['seasoning']) != ck_sea):
        count+=1

print(f'duplicated : {count}')
```
