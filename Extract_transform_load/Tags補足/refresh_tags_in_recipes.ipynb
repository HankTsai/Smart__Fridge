#### 新增tag標籤到食譜中

```python
import time
import json
```

```python
with open('./foodBank/recipe0925_V1.json', 'r', encoding='utf-8') as f:
    content = json.loads(f.read())
```

```python
with open('./foodBank/new_tag_with_catagories.json', 'r', encoding='utf-8') as f:
    tags = json.loads(f.read())
```

```python
tags[1]
```

```python
# 轉成字典方便配對
big_dict = dict()
for tag in tags:
    for urls in list(tag.values()):
        for url in urls:
            if url not in big_dict:
                big_dict[url] = [list(tag.keys())[0]]
            else:
                big_dict[url] += [list(tag.keys())[0]]
```

```python
len(big_dict.keys())
```

```python
big_dict['https://icook.tw//recipes/178945']
```

```python
count = 0
for k in content:
    url = 'https://icook.tw//recipes/' + k['url'][25:]
    if url in big_dict:
        count += 1
        k['tags'] += big_dict[url]
print(count)
```

```python
for k in content:
    if k['url'] == 'https://icook.tw/recipes/221396':
        print(k['tags'])
        print(k)
```

```python
# write into new json
with open('./foodBank/recipe0925_V2.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(content))
```

```python
all_tags = list()
for i in tags:
#     print(list(i.keys())[0])
    all_tags.append(list(i.keys())[0])
```

```python
all_tags_str = ','.join(all_tags)
```

```python
# write into a list of all tags
with open('./foodBank/tags.txt', 'w', encoding='utf-8') as f:
    f.writelines(all_tags_str)
```
