```python
import json
import pandas as pd
import time
```

```python
with open(r'C:\Users\Big data\PycharmProjects\collaborative_filtering\dataset\recipe0928_V5.json','r')as f:
    content = json.load(f)
```

```python
content[:5]
```

```python
# 將食譜名稱及食材資訊轉成list以便轉成dataframe
df_list = list()
for k in content:
    each_list = list()
    ing_list = list()
    for i in k['ingredient']:
        ing_list.append(i[0])
        
    ing_str = ','.join(ing_list)
    each_list.append(k['recipe'])
    each_list.append(ing_str)
    df_list.append(each_list)
```

```python
df_list[:100]
```

```python
# 轉成dataframe
df = pd.DataFrame(df_list)
```

```python
df
```

```python
# 加上column name
df.columns = ['recipe','ingredient']
```

```python
df
```

```python
# 取出食材做成特徵矩陣
dummies = df['ingredient'].str.get_dummies(',')
```

```python
dummies
```

```python
# 把食材特徵矩陣跟原食譜矩陣結合
recipe_vec = pd.concat([df, dummies], axis=1)
recipe_vec.drop('ingredient',axis=1,inplace=True)
```

```python
recipe_vec.set_index('recipe',inplace=True)
```

```python
recipe_vec[:10]
```

```python
# calaulate cosine simularity
from math import sqrt
def cosine_sim(a,b):
    sum_xy = 0
    sum_xx = 0
    sum_yy = 0
    for i in range(len(recipe_vec.columns)):
        sum_xy += a[i] * b[i]
        sum_xx += a[i] * b[i]
        sum_yy += b[i] * b[i]
    try:
        cosine_simularity = sum_xy / (sqrt(sum_xx) * sqrt(sum_yy))
    except:
        return "Error,can't calculate the simularity"
    return cosine_simularity
```

```python
# 比較兩者的相似度
a = recipe_vec.loc['也紹興腐乳肉']
b = recipe_vec.loc['肉燥飯']
cosine_sim(a,b)
```
